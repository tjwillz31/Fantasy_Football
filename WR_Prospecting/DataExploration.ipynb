## Data Exploration

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('prospects_with_FF_stats.csv')
```

```python
# filter the player pool to 2018 and earlier
# 2018 has only had 2 seasons to become a 3 Year Hit, but the guys that most look like hits already have, so count it
data3y = data[data['Draft_Year'] < 2019]
```

```python
data3y['3Year_Hit'].value_counts()
```

### Draft Factors

```python
# explore 3 Year Hits by draft capital
ax = sns.violinplot(x='3Year_Hit', y='Overall_Pick', data=data3y)
```

```python
ax = sns.countplot(x='Draft_Round', hue='3Year_Hit', data=data3y)
```

```python
pd.crosstab(index=data3y['3Year_Hit'], columns=data3y['Draft_Round'])
```

```python
# explore 3 Year Hits by early declare
ax = sns.countplot(x='Early_Declare', hue='3Year_Hit', data=data3y)
```

```python
pd.crosstab(index=data3y['3Year_Hit'], columns=data3y['Early_Declare'])
```

```python
# combine both
g = sns.catplot(x='Draft_Round', hue='3Year_Hit', col='Early_Declare',
                data=data3y, kind="count");
```

```python
pd.crosstab(index=[data3y['Early_Declare'], data3y['3Year_Hit']], columns=data3y['Draft_Round'])
```

```python
ax = sns.violinplot(x='3Year_Hit', y='Draft_Age', data=data3y)
```
